**What does each of the column means?**

Columns 'GAD1', 'GAD2', 'GAD3', 'GAD4', 'GAD5', 'GAD6', 'GAD7' corresponds to diffent questions asked in the General Anxiety Disorder (GAD) questionnaire.The GAD questionnaire consisted of 7 questions to which your answers must be selected.The following are the options with  values they represent in the dataset:

* Not at all [0]

* Several days [1]

* More than half the days[2]

* Nearly every day[3]  


* **GAD1** - Feeling nervous,anxious or on edge?
* **GAD2** - Not being able to stop or control worrying?
* **GAD3** - Worrying too much about different things?
* **GAD4** - Triuble relaxing?
* **GAD5** - Being so restless that it is hard to sit still
* **GAD6** - Becoming easily annoyed or irritable?
* **GAD7** - Feeling afraid as if something awful might happen?
* **GADE** - This column asks affect of gaming in work. If it was difficult,some what difficult or not difficult at all etc.. to fill out the questionnaire.  

Columns 'SWL1','SWL2', 'SWL3', 'SWL4', 'SWL5' corresponds to questions of Satisfaction With Life (SWL) questionnaires.The SWL questionnaire consisted of 5 questions to which your answers must must be selected.The following are the options with the values they represent in the dataset:

* Strongly agree[7]

* Agree[6]

* Slightly agree[5]

* Neither agree nor disagree[4]

* Slightly disagree [3]

* Disagree[2]

* Strongly disagree[1]  


* **SWL1** - In most ways my life is close to my ideal
* **SWL2** - The conditions of my life are excellent
* **SWL3** - I am satisfied with my life
* **SWL4** - So far I have gotten the important things I want in life
* **SWL5** - If I could live my life over, I would change almost nothing
* **Game** - Name of the game they play
* **Platform** - Name of platform used for gaming
* **Hours** - No. of hours engaged in gaming per week
* **earnings** - Whether they earn money with these games
* **whyplay** - Reason for gaming
* **League** - League in which currently playing
* **highestleague** - highest league to be ever played
* **streams** - No. of hours engaged in streaming per week
* **Gender** - Gender of gamer
* **Age** - Age of gamer
* **Work** - profession of gamer
* **Degree** - Highest qualification of the gamer
* **Birthplace** - Contains Bithplace of gamers
* **Residence** - Contains current residence of the gamer
* **Reference** - The site of reference
* **Playstyle** - Whethe the gamer is single player or multiplayer
* **GAD_T** - Total of all GAD questionnaire answers scores
* **SWL_T** - Total of all SWL questionnaire answers scores
* **Residence_ISO3** - Birthplace after encoding
* **Birthplace_ISO3** - Current Residence after encoding

### Preprocessing



```python
df.isnull().sum()
```

```python
# Checking for missing values in categorical features
cat_with_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']
cat_with_nan
```

```python
for feature in cat_with_nan:
    print(feature, np.round(df[feature].isnull().mean(),3), '% of missing values')
```

```python
# handling missing values with most frequent value
for feature in cat_with_nan:
    most_frequent_cat = df[feature].value_counts().index[0]
    df[feature].fillna(most_frequent_cat, inplace = True)
```

```python
# Checking for missing values in numerical features
num_with_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

for feature in num_with_nan:
    print(feature, np.round(df[feature].isnull().mean(),3), '% of missing values')
```

```python
# handling the missing values with median
for feature in num_with_nan:
    med = df[feature].median()
    df[feature].fillna(med, inplace=True)
```

```python
df.isnull().sum()
```

```python
# plotting skewness
import warnings
warnings.filterwarnings('ignore')
for feature in continuous_features:
    data = df.copy()
    
    sns.histplot(df[feature],  binrange=(-2, 500))
    plt.show()
```

since both are right skewed, we need to perform log transformation

```python
# Applying log transformation to continuous numerical features
for feature in continuous_features:
    df[feature] = np.log(df[feature]+1)
```

```python
# plotting boxplot for outlier detection
for feature in continuous_features:
  plt.figure(figsize=(16,8))
  plt.subplot(2,2,1)
  sns.histplot(df[feature], kde=True)

  plt.subplot(2,2,2)
  sns.boxplot(data = df[feature])
  plt.show()
```

```python
# finding outliers
for feature in continuous_features:
        Q1 = np.percentile(df[feature], 25)
        Q3 =np.percentile(df[feature], 75)
        IQR = Q3 - Q1
        low_limit = Q1-1.5*IQR
        upr_limit = Q3+1.5*IQR
        print(feature, 'lower limit is:',low_limit)
        print(feature, 'Upper limit is:',upr_limit)
```

```python
outlier=[]
for feature in continuous_features:
  for x in df[feature]:
    if ((x>upr_limit) or (x<low_limit)):
        outlier.append(x)
```

```python
from sklearn.datasets import make_multilabel_classification
from sklearn.model_selection import train_test_split

# Generate a synthetic multi-label classification dataset with 1000 samples, 21 features, 2 classes, and 2 labels per instance, using a random state of 1 for reproducibility
X, y = make_multilabel_classification(n_samples=1000, n_features=21, n_classes=2, n_labels=2, random_state=1)

# Split the data into training and testing sets with a 75/25 split and a random state of 42 for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

**Model Selection**

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.multioutput import MultiOutputClassifier

rf = RandomForestClassifier(n_estimators=1000, random_state=42, max_depth=8)
multi_rf = MultiOutputClassifier(rf)

multi_rf.fit(X_train, y_train)
y_pred_rf = multi_rf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred_rf)
label_names = ['Anxiety_level', 'Life_satisfaction']
print("Random Forest Classifier")
print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred_rf, target_names=label_names))
```

```python
# plotting confusion matrix
from sklearn.metrics import multilabel_confusion_matrix, ConfusionMatrixDisplay
cm = multilabel_confusion_matrix(y_test, y_pred_rf)
disp = ConfusionMatrixDisplay(cm[0], display_labels=[0, 1])
disp.plot(cmap='Blues', values_format='.0f')
disp.ax_.set_title('Confusion matrix for Anxiety_level')

disp = ConfusionMatrixDisplay(cm[1], display_labels=[0, 1])
disp.plot(cmap='Blues', values_format='.0f')
disp.ax_.set_title('Confusion matrix for Life_satisfaction')
```

```python
# Iterative Stratification Cross validation
from skmultilearn.model_selection import IterativeStratification

# Create an IterativeStratification object for k-fold cross-validation
k = 10
iterative_validator = IterativeStratification(n_splits=k, order=1, sample_distribution_per_fold=None)

# Perform k-fold cross-validation and get the accuracy scores
accuracy_scores = []
for train_index, test_index in iterative_validator.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    multi_rf.fit(X_train, y_train)
    y_pred_rf = multi_rf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred_rf)
    accuracy_scores.append(accuracy)

mean_accuracy = np.mean(accuracy_scores)
print("Random Forest Classifier - Iterative Stratification cross-validation")
print("Mean accuracy:", mean_accuracy)
```

```python
# Leave-One-Group-Out cross-validation
from sklearn.model_selection import LeaveOneGroupOut

# Create an array of groups
groups = [i // 10 for i in range(X.shape[0])]

# Create a Leave-One-Group-Out object for cross-validation
logo_validator = LeaveOneGroupOut()

# Perform cross-validation and get the accuracy scores
accuracy_scores = []
for train_index, test_index in logo_validator.split(X, y, groups):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    multi_rf.fit(X_train, y_train)
    y_pred_rf = multi_rf.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred_rf)
    accuracy_scores.append(accuracy)

print("Random Forest Classifier - Leave-One-Group-Out cross-validation")
print("Average Accuracy:", sum(accuracy_scores)/len(accuracy_scores))
```

Random Forest algorithm has achieved an accuracy of 84%, and after cross-validation, the accuracy has improved slightly to a range of **84.6%** to **85.1%** using different techniques.
